```python
X = mobile_price_final_data.drop(["Model", "Price (TK)"], axis = 1)

X.head()
```

```python
XD = pd.get_dummies(X)

XD.head()
```

```python
XD.columns
```

```python
Y = mobile_price_final_data["Price (TK)"]

Y[:10]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(XD, Y, train_size = 0.8)

X_train.shape, X_test.shape
```

## Cross validation score

```python
def Cross_validation(model, X, Y):

  from sklearn.model_selection import cross_val_score
  c = cross_val_score(model, X, Y, cv = 5)

  print("\nCross Validation Scores: ")
  for i in range(len(c)):
    print(f"For Split {i+1}: {c[i]*100:.2f}%")
  print("\n")
  
```

## Evalutionary parameters

```python
def Evalution_metrics(M_name, Score, Y_test, Y_preds):
  
  print(f"{M_name}'s Evalution: \n")

  # Accuracy
  print(f"Accuracy: {Score*100:.2f}%")

  # R^2 Score
  from sklearn.metrics import r2_score
  r2 = r2_score(Y_test, Y_preds)
  print(f"R^2 Score: {r2:.2f}")

  # MAE
  from sklearn.metrics import mean_absolute_error
  mae = mean_absolute_error(Y_test, Y_preds)
  print(f"Mean Absolute error: {mae:.2f}")

  # MSE
  from sklearn.metrics import mean_squared_error
  mse = mean_squared_error(Y_test, Y_preds)
  print(f"Mean Squared error: {mse:.2f}")
  
```

## Shows model's true, predicted, difference between these & it's squared values.

```python

def View_df(Y_test, Y_preds):

  df = pd.DataFrame(
      {
          "True values" : Y_test,
          "Predicted values" : Y_preds
      }
  )

  df["Differences"] = np.absolute(df["True values"] - df["Predicted values"])

  df["Squared"] = np.square(df["Differences"])

  return df
    
```

# **Applying models & checking their performances.**

# RidgeRegressor

```python
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

model = make_pipeline(StandardScaler(), Ridge(alpha = 1.0))

model.fit(X_train, Y_train)

Score = model.score(X_test, Y_test)

Y_preds = model.predict(X_test)

Evalution_metrics("RidgeRegressor",Score, Y_test, Y_preds)

Cross_validation(model, XD, Y)

View_df(Y_test, Y_preds).head(10)
```

# RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(max_depth=2, random_state=0)

model.fit(X_train, Y_train)

Y_preds = model.predict(X_test)

Score = model.score(X_test, Y_test)

Evalution_metrics("RandomForestRegressor",Score, Y_test, Y_preds)

Cross_validation(model, XD, Y)

View_df(Y_test, Y_preds).head(10)
```

## GradientBoostingRegressor

```python
from sklearn.ensemble import GradientBoostingRegressor

model = GradientBoostingRegressor(n_estimators = 100, learning_rate = 0.1, max_depth = 1, random_state = 0)

model.fit(X_train, Y_train)

Y_preds = model.predict(X_test)

Score = model.score(X_test, Y_test)

Evalution_metrics("GradientBoostingRegressor",Score, Y_test, Y_preds)

Cross_validation(model, XD, Y)

View_df(Y_test, Y_preds).head(10)
```

# DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(random_state = 0)

model.fit(X_train, Y_train)

Y_preds = model.predict(X_test)

Score = model.score(X_test, Y_test)

Evalution_metrics("DecisionTreeRegressor",Score, Y_test, Y_preds)

Cross_validation(model, XD, Y)

View_df(Y_test, Y_preds).head(10)
```
